```python
#Bank App

from random import randint
class BankApp:
    acc_no = int()  #Account Number
    pas = str()
    bal = int()
        
    def display(self):
        choice = int(input("1.Login\n2.SignUp\n3.Exit\n"))
        if choice == 3:
            print("Thanks for Visiting")
        elif choice == 2:
            self.signup()
        elif choice == 1:
            self.login()   
        
    def signup(self):
        name = input("Name: ")
        self.pas = input("Password: ")
        self.bal = float(input("Balance: "))
        print("Your Account No. is :",end=' ')
        self.acc_no = self.acc()   # Calls the account number generator fun
        print(self.acc_no)
        self.display()
        
    
    def login(self):
        acc = int(input("Please type Account number: "))
        pas = str(input("Please type Password: "))
        if acc == self.acc_no and pas == self.pas:
            print("You are Logged in")
            self.internal()
        else:
            print("Wrong Account number and Password")
            self.display()
        
    def internal(self):
        print("Enter a choice: ")
        ch=int(input("1.Credit\n2.Debit\n3.Check Balance\n4.Logout\n"))
        
        if ch == 1: 
            print("Present Balance: ",self.bal)
            amt = int(input("Credit Amount: "))
            self.bal = self.bal+amt
            print("Credited Succesfully \nBalance: ",self.bal)
            self.internal()
            
        elif ch == 2:
            if self.bal>0:
                print("Your Balance: ",self.bal)
                amt = int(input("Amount to Debited: "))
                self.bal = self.bal-amt
                print("Debited succesfully \nBalance: ",self.bal)
                self.internal()
        
        elif ch == 3:
            print("Balance: ",self.bal)
            self.internal()
        
        elif ch == 4:
            self.display()
            
        else:
            print("Wrong choice try again")
            self.internal()
            
    
    def acc(self):       # Account No. generator
        import random
        l = list()
        for i in range(0,2):
            l.append(random.randint(10,90))
        a = [str(i) for i in l]
        x = int("".join(a))
        return x
    
B = BankApp()
B.display()
```
